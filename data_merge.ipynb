```python
import pandas as pd
import os
from zipfile import ZipFile
import glob
import matplotlib.pyplot as plt
```

```python
cols = ["Perimetre", "Nature", "Date", "Heures", "y", "Thermique", "Nucleaire", "Eolien", "Solaire", "Hydraulique", "Pompage", "Bioenergies", "Ech. physiques"]
```

```python
path = "data"
all_files = glob.glob(os.path.join(path, "*.xls"))
df = pd.concat((pd.read_csv(f, names = cols,sep='\t', skiprows=1, index_col=False)) for f in all_files)
```

```python
# df.to_csv("all_data.csv")
df.head()
```

```python
df = df.dropna(subset = ['y'])
df
```

```python
weather_df = pd.read_csv("data\donnees-synop-essentielles-omm.csv",sep=";")
```

```python
IDF_df = weather_df[weather_df["region (name)"] == "Île-de-France"]
IDF_df.head()
```

Taking only data between 2016-2020 and only temperature and humidity data

```python
yrs = ['2016','2017','2018','2019','2020']

IDF_df = IDF_df[IDF_df["Date"]>="2016-01-01T00:00:00+01:00"]
IDF_df = IDF_df[['Date','Température','Humidité','Pression au niveau mer']]
IDF_df['Heures'] = [d[11:16] for d in IDF_df['Date']]
IDF_df['Date'] = [d[:10] for d in IDF_df['Date']]     
IDF_df.head()   
```

Merging and linearly interpolating data to match consumption data time steps

```python
# df_temp_hum= pd.merge(df,IDF_df, on = ['Date','Heures'])
df_temp_hum= pd.merge(df,IDF_df,how='left', on = ['Date','Heures'])
df_temp_hum = df_temp_hum[['Date','Heures','y','Température','Humidité','Pression au niveau mer']]
df_temp_hum = df_temp_hum.interpolate()
df_temp_hum = df_temp_hum.dropna()
df_temp_hum
```

Adding Holidays column

```python
holidays = ['01-01','05-09','11-01','04-01','05-20','11-09','05-01','07-14','12-25','05-08','08-15']

df_temp_hum['Holiday'] = [1 if d[5:] in holidays else 0 for d in df_temp_hum['Date']]
```

Renaming columns and saving df to CSV

```python
df_temp_hum = df_temp_hum[['Date','Heures','y','Température','Humidité','Pression au niveau mer','Holiday']]
df_temp_hum = df_temp_hum.rename(columns={'Heures':'Hour','Température':'Temperature','Humidité':'Humidity','Pression au niveau mer':'Sealevel Pressure'})
df_temp_hum.to_csv("data_temp_hum_hol.csv")
df_temp_hum
```
